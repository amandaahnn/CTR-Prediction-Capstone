### Exploration of each variable

For the sake of easy EDA and graphic visualization, I will be creating a temporary dataframe (called data_graph) with less data, so my code will run faster. 

```python
data_graph1 = data[data['click']==0]
data_graph2 = data[data['click']==1]
print(data_graph1.shape, data_graph2.shape)
```

```python
data_graph1 = data_graph1.sample(frac=0.01, random_state=42)
data_graph1.shape
```

```python
data_graph2 = data_graph2.sample(frac=0.03, random_state=42)
data_graph2.shape
```

```python
data_graph = pd.concat([data_graph1, data_graph2], axis = 0, ignore_index=True)
data_graph['click'].value_counts()
```

We will test a variety of plots to explore different features in the data and see if we could find any relationships between them. We will use pandas boxplot, seaborn pairplot, and seaborn catplots. 

```python
data_graph.boxplot(by='click', column ='banner_pos', grid=False, rot=45, fontsize=15)
```

```python
sns.pairplot(data_graph)
```

From this pairplot, features C14 and C17 seem like the the only ones that have some sort of positive correlation between each other. These features could hold some importance in future analysis, so we will make sure to note this. In order to analyze the categorical features like site category and site domain, we will be plotting each feature separately with banner position, while differentiating between a click and no click through the hue parameter. 

```python
sns.stripplot(data=data_graph, x="banner_pos", y="site_category", hue="click")
plt.xlabel('Banner Position', fontsize=15)
plt.ylabel('Site Category', fontsize=15)
plt.show()
```

```python
sns.stripplot(data=data_graph, x="banner_pos", y="C17", hue="click")
plt.xlabel('Banner Position', fontsize=15)
plt.ylabel('C17', fontsize=15)
plt.show()
```

```python
sns.stripplot(data=data_graph, x="banner_pos", y="C14", hue="click")
plt.xlabel('Banner Position', fontsize=15)
plt.ylabel('C14', fontsize=15)
plt.show()
```

```python
sns.stripplot(data=data_graph, x="banner_pos", y="C21", hue="click")
plt.xlabel('Banner Position', fontsize=15)
plt.ylabel('C21', fontsize=15)
plt.show()
```

```python
sns.stripplot(data=data_graph, x="click", y="banner_pos", hue='click')
plt.xlabel('Click', fontsize=15)
plt.ylabel('Banner Position', fontsize=15)
plt.show()
```

```python
sns.stripplot(data=data_graph, x="banner_pos", y="app_domain", hue="click")
plt.xlabel('Banner Position', fontsize=15)
plt.ylabel('App Domain', fontsize=15)
plt.show()
```

```python
sns.stripplot(data=data_graph, x="banner_pos", y="app_category", hue="click")
plt.xlabel('Banner Position', fontsize=15)
plt.ylabel('App Category', fontsize=15)
plt.show()
```

### Modeling (formulate null hypothesis and alternative hypothesis)

Null: the prediction of a click is due to chance and not due to the positioning of the ad on the site/app/device

Alternative: the prediction of a click is not due to chance and it has something to do with the positioning of the ad

Going forward, I believe these null and alternative hypotheses would be a good start to our analysis. Even if we fail to reject the null, then the analysis would be informative because then we would know the banner position has no effect on click through rate. However, it might be a better idea to have a null hypothesis that we can reject so that we can find something significant, but it is difficult to make predictions when most of the features are hidden features.
